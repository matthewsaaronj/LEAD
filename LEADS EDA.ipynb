### Time Series Analysis on Meter Readings

```python
# resample data to weekly level
weekly_data = (
    train
    .groupby('building_id')
    .resample('W')
    .mean()
    .drop('building_id', axis=1)
    .reset_index()
)

weekly_data
```

```python
hv.extension('bokeh')
plot = hv.Overlay()

for building_id in weekly_data['building_id'].unique():
    data = weekly_data[weekly_data['building_id'] == building_id]
    curve = hv.Curve((data['timestamp'], data['meter_reading']), 'timestamp', 'meter_reading')
    plot *= curve

plot.opts(width=600, height=400, legend_position='right')
```

### How far can we get with a simple approach using distributions?
This approach will help set a baseline before we move to more complex solutions. 

```python
import pandas as pd

class MeterReadingOutlierDetector:
    def __init__(self, std_dev_threshold):
        self.std_dev_threshold = std_dev_threshold
        self.means = None
        self.stds = None

    def fit(self, df):
        # Calculate the mean and standard deviation for each building_id group
        grouped = df.groupby('building_id')['meter_reading']
        self.means = grouped.mean()
        self.stds = grouped.std()

    def transform(self, df):
        # Create a new column to store the outlier flag
        outlier_flags = []
        
        for _, row in df.iterrows():
            building_id = row['building_id']
            meter_reading = row['meter_reading']
            mean = self.means[building_id]
            std = self.stds[building_id]
            
            # Check if the meter reading is outside the specified standard deviation range
            if abs(meter_reading - mean) > self.std_dev_threshold * std:
                outlier_flags.append(1)
            else:
                outlier_flags.append(0)
        
        return outlier_flags

    def fit_transform(self, df):
        self.fit(df)
        return self.transform(df)

# Example usage
data = {
    'building_id': [1, 1, 1, 2, 2, 2, 3, 3, 3],
    'meter_reading': [100, 300, 110, 200, 205, 210, 300, 305, 310]
}

```

```python
detector = MeterReadingOutlierDetector(std_dev_threshold=2)

train['pred_anomaly'] = detector.fit_transform(train)
train
```

```python
roc_auc_score(train.anomaly, train.pred_anomaly)
```
